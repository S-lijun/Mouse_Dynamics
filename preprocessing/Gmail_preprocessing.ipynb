```python
import os
import shutil
import pandas as pd
from tqdm import tqdm

# === directory path ===
SOURCE_ROOT = '../Data/Gmail-dataset'
TARGET_ROOT = '../Data/Gmail'
os.makedirs(TARGET_ROOT, exist_ok=True)

# === skip user original code ===
skip_user_ids = {3, 4, 5, 6, 9, 13, 21, 25, 26, 34, 38, 42}

# === original code mapping to 0~42 ===
original_id_to_folder = {
    0: '2884292', 1: '9911071', 2: '1513190', 3: '1067144', 4: '7495794',
    5: '7921384', 6: '6485798', 7: '2330116', 8: '7030676', 9: '1514461',
    10: '2164619', 11: '9031166', 12: '6755264', 13: '6425139', 14: '6973075',
    15: '5125125', 16: '8941269', 17: '7164913', 18: '4064487', 19: '9609325',
    20: '6086694', 21: '3604291', 22: '7547835', 23: '2451411', 24: '2481013',
    25: '2057188', 26: '4906875', 27: '5389326', 28: '4376779', 29: '3667465',
    30: '8139946', 31: '9604161', 32: '5563756', 33: '8058937', 34: '7882967',
    35: '9779049', 36: '7462635', 37: '9800000', 38: '5987580', 39: '7642950',
    40: '7058298', 41: '6414470', 42: '9274054'
}

# === ===
valid_users = [(k, v) for k, v in original_id_to_folder.items() if k not in skip_user_ids]
remap_table = {folder: f'user{new_idx}' for new_idx, (k, folder) in enumerate(valid_users)}

# === copy data ===
for original_folder, user_folder in tqdm(remap_table.items()):
    source_user_path = os.path.join(SOURCE_ROOT, original_folder)
    target_user_path = os.path.join(TARGET_ROOT, user_folder)
    os.makedirs(target_user_path, exist_ok=True)

    if not os.path.isdir(source_user_path):
        print(f"Can't find directory: {source_user_path}")
        continue

    for fname in os.listdir(source_user_path):
        if fname.endswith('.csv'):
            shutil.copy(
                os.path.join(source_user_path, fname),
                os.path.join(target_user_path, fname)
            )

print("Done")
```
